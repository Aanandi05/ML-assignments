```python
from PIL import Image
import numpy as np

def img_to_array(path):
    # Load the image
    img = Image.open(path)

    # Convert image to numpy array
    img_array = np.array(img)

    # Determine if the image is RGB or Grayscale
    if len(img_array.shape) == 3:  # RGB image
        file_name = path.split('.')[0] + '_rgb.txt'
        # Flatten the array and save it
        with open(file_name, 'w') as f:
            for row in img_array:
                np.savetxt(f, row, fmt='%d', newline=' ')
                f.write('\n')
    elif len(img_array.shape) == 2:  # Grayscale image
        file_name = path.split('.')[0] + '_gray.txt'
        # Flatten the array and save it
        with open(file_name, 'w') as f:
            np.savetxt(f, img_array, fmt='%d', newline=' ')
            f.write('\n')
    else:
        raise ValueError("Unsupported image format")

    print(f"Image data has been saved to {file_name}")


```

1. Discrete vs. Continuous Data:
Discrete Data: Countable values, often integers (e.g., number of children, cars owned).
Continuous Data: Measurable values, can take any value within a range (e.g., age, income).
2. Levels of Measurement:
Nominal: Categories with no inherent order (e.g., occupation, gender).
Ordinal: Ordered categories, but the difference between them isnâ€™t meaningful (e.g., education level).
Interval: Ordered with meaningful differences, but no true zero (e.g., birthdate, temperature in Celsius).
Ratio: Like interval, but with a true zero, allowing for meaningful ratios (e.g., income, weight).

```python
#Using drop
print("Shape before dropping nulls:", df.shape)

df = df.dropna()

print("Shape after dropping nulls:", df.shape)
```

```python
df= df.dropna(axis=1, how='all')

df_num= df.select_dtypes(include='number')
df_num.head() #all the numeric columns
df[df_num.columns]= df_num.fillna(df_num.mean()) #filling the null values in numeric columns with mean
df.head(10)

df_cat= df.select_dtypes(include= 'object')
df_cat.head() #all the categorical columns
df[df_cat.columns]= df_cat.fillna(df_cat.mode()) #filling the null values in categorical columns with mode
df.head(20)

```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df_num = df.select_dtypes(include=['number'])
scaler = MinMaxScaler()
df_norm = pd.DataFrame(scaler.fit_transform(df_num), columns=df_num.columns)

print(df_norm.head(20))
```

```python

df_selected = df[selected_columns]

df_selected = df_selected.dropna()

df_transformed = pd.get_dummies(df_selected, columns=['Gender','MaritalStatus','HomeOwnerFlag','YearlyIncome'])
print(df_transformed.head())
```

```python
import pandas as pd
from sklearn.metrics import jaccard_score
from sklearn.metrics.pairwise import cosine_similarity

row1 = df_transformed.iloc[0].values
row2 = df_transformed.iloc[1].values

# Simple Matching Coefficient (SMC)
simple_matching = (row1 == row2).mean()

# Jaccard Similarity for Binary Data
try:
    jaccard_similarity = jaccard_score(row1, row2, average='binary')
except ValueError:
    print("Jaccard similarity can only be used with binary data.")

# Cosine Similarity for Numeric Data
cosine_similarity_value = cosine_similarity([row1], [row2])[0][0]

# Output the similarity results
print("Simple Matching Coefficient (SMC):", simple_matching)
print("Cosine Similarity:", cosine_similarity_value)
```
